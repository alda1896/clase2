```python
from numpy import linspace, sin, cos

x= linspace (0,10,100)
y= sin(x)
plot(x,y)
show
```

```python
from pylab import ylim, xlim, grid

x= linspace (0,10,100)
y= sin(x)
plot(x,y)
ylim (-1.1,1.1)
xlim (0., 10)
grid ("on")
show

```

```python
from pylab import xlabel, ylabel

x= linspace (0,10,100)
y= sin(x)
plot(x,y)
ylabel ("eje y")
xlabel ("eje x")
grid ("on")
show
```
